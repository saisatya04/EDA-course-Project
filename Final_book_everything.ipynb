```python
data = pd.read_csv("GYTS4(2).csv")
data.head()
```

```python
regions = {
    'North': ['Jammu and Kashmir', 'Himachal Pradesh', 'Punjab', 'Haryana', 'Uttarakhand', 'Delhi', 'Uttar Pradesh', 'Chandigarh'],
    'South': ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana', 'Puducherry', 'Lakshadweep'],
    'East': ['Bihar', 'Jharkhand', 'Odisha', 'West Bengal'],
    'West': ['Goa', 'Gujarat', 'Maharashtra', 'Rajasthan', 'Dadra and Nagar Haveli and Daman and Diu'],
    'Central': ['Madhya Pradesh', 'Chhattisgarh'],
    'Northeast': ['Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Sikkim', 'Tripura'],
    'Union Territories': ['Andaman and Nicobar Islands', 'Chandigarh', 'Dadra and Nagar Haweli', 'Daman and Diu', 'Lakshadweep', 'Delhi', 'Puducherry']
}

def cluster_states_by_region(states):
    clustered_states = {region: [] for region in regions.keys()}
    
    for state in states:
        found = False
        for region, region_states in regions.items():
            if state in region_states:
                clustered_states[region].append(state)
                found = True
                break
        if not found:
            clustered_states.setdefault('Unknown', []).append(state) 
    
    return clustered_states

states_to_cluster = data["State/UT"]
clustered_states = cluster_states_by_region(states_to_cluster)

for region, states in clustered_states.items():
    if states:
        print(f"{region}: {', ' if states else ''}{', '.join(states)}")
```

```python
state_to_region = {state: region for region, states in regions.items() for state in states}

def get_region(state):
    return state_to_region.get(state, 'Unknown')

data['Region'] = data['State/UT'].apply(get_region)

print(data)
```

```python
# Identify numeric columns
numeric_cols = data.select_dtypes(include='number').columns

# Merge rows with respect to states and calculate the mean for numeric columns
merged_data = data.groupby('State/UT')[numeric_cols].mean().reset_index()

# Display the merged data
print(merged_data.head())

# Plotting
plt.figure(figsize=(18, 10))  # Increase figure size for better clarity

# Create bar plot with a professional color palette
bar_plot = sns.barplot(data=merged_data, x='State/UT', y='Current tobacco users (%)', ci=None, palette='viridis')

# Display the percentage on top of each bar with smaller font size
for index, row in merged_data.iterrows():
    bar_plot.text(index, row['Current tobacco users (%)'] + 0.5, f"{row['Current tobacco users (%)']:.2f}%", 
                  color='black', ha="center", fontsize=10)

# Customize plot
plt.xticks(rotation=90, fontsize=12)  # Rotate x labels for better readability
plt.yticks(fontsize=12)
plt.title('Current Tobacco Users by State/UT', fontsize=16)
plt.xlabel('State/UT', fontsize=14)
plt.ylabel('Current Tobacco Users (%)', fontsize=14)

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Sort merged_data by 'Current tobacco users (%)' in descending order
top_3_states = merged_data.sort_values(by='Current tobacco users (%)', ascending=False).head(3)

# Display the top 3 states
print("Top 3 States with Highest Percentage of Current Tobacco Users:")
print(top_3_states[['State/UT', 'Current tobacco users (%)']])

# Optionally, plot the top 3 states on the bar plot with a different color
plt.figure(figsize=(18, 10))  # Increase figure size for better clarity

# Create bar plot with a professional color palette
bar_plot = sns.barplot(data=merged_data, x='State/UT', y='Current tobacco users (%)', ci=None, palette='viridis')

# Highlight the top 3 states with a different color
for index, row in top_3_states.iterrows():
    bar_plot.patches[index].set_facecolor('orange')  # Change color for top 3 states

# Display the percentage on top of each bar with smaller font size
for index, row in merged_data.iterrows():
    bar_plot.text(index, row['Current tobacco users (%)'] + 0.5, f"{row['Current tobacco users (%)']:.2f}%", 
                  color='black', ha="center", fontsize=10)

# Customize plot
plt.xticks(rotation=90, fontsize=12)  # Rotate x labels for better readability
plt.yticks(fontsize=12)
plt.title('Current Tobacco Users by State/UT', fontsize=16)
plt.xlabel('State/UT', fontsize=14)
plt.ylabel('Current Tobacco Users (%)', fontsize=14)

# Add grid lines for better readability
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
